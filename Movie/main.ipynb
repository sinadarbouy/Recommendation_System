```python
import pandas as pd
movies = pd.read_csv("movie_genres.csv", sep=",", index_col=0)
movies.head()
```

```python
movies['movieId'] = np.arange(len(movies))+1
movies.head()
```

```python
users = pd.read_csv("user_reviews.csv", sep=",", index_col=0)
users.head()
```

```python
users.duplicated(["User"]).sum()
users.describe()
```

```python
users_T = users.T
users_T
```

```python
data = {
  "userId": [],
  "movieId": [],
  "rating":[]
}

row = 0
df = pd.DataFrame(data)
for i in range(600):
    print(i)
    movieid = 1
    for x in users_T[users_T.columns[i]][1:]:
        new_row = {'userId':i+1, 'movieId':movieid, 'rating':x}
        df = df.append(new_row, ignore_index=True)
        movieid += 1
        row += 1
df
```

```python
df.to_csv('MovieRatingData.csv', index=False)
```

```python
df = pd.read_csv("MovieRatingData.csv", sep=",")
```

```python
import numpy as np
```

```python
df["rating"].replace(0,np.nan,inplace=True)
```

```python

data_withMovieName = pd.merge(df,movies, on='movieId')
data_withMovieName = data_withMovieName[["userId","movieId","movie_title","rating"]]
data_withMovieName.head()
```

```python
data_withMovieName.groupby('movie_title')['rating'].mean().sort_values(ascending=False).head()
```

```python
data_withMovieName.groupby('movie_title')['rating'].count().sort_values(ascending=False).head()
```

```python
mean_movie_table=data_withMovieName.groupby('movie_title')['rating'].mean().sort_values(ascending=False) 
count_movie_table=data_withMovieName.groupby('movie_title')['rating'].count().sort_values(ascending=False)
Movie_table = pd.merge(mean_movie_table,count_movie_table, on='movie_title')
Movie_table.columns =['mean', 'count']
Movie_table.sort_values('count', ascending = False).head(10)
```

```python
movie_pivot = pd.pivot_table(data_withMovieName,index ='userId',columns ='movie_title', values ='rating')
movie_pivot
```

```python
Furious7_data = movie_pivot['Furious 7']
Furious7_data
```

```python
similar_to_Furious7 = movie_pivot.corrwith(Furious7_data) # pearson, Kendall, spearman
corr_Furious7 = pd.DataFrame(similar_to_Furious7, columns =['Correlation'])
corr_Furious7.dropna(inplace = True)
 
corr_Furious7.head()
```

```python
corr_Furious7.sort_values('Correlation', ascending = False).head(10)
corr_Furious7 = corr_Furious7.join(Movie_table['count'])
```

```python
corr_Furious7
corr_Furious7[corr_Furious7['count']>12].sort_values(['Correlation','count'], ascending = (False,False)).head()
```
